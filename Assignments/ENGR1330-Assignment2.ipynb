```python
# B)
score_by_month = {"Sally": {"Jan": 670, "Feb": 689, "Mar": 685},
                  "Dominic": {"Jan": 720, "Feb": 714, "Mar": 725},
                  "Herbert": {"Jan": 801, "Feb": 799, "Mar": 804},
                  "Wendy": {"Jan": 745, "Feb": 741, "Mar": 751}
                 }
# it might be more efficient to have the scores saved in a list, however, the data structure would be less self explainitory

individual_average = dict()

for person in score_by_month:
    # calculate each persons' average monthly score
    score_total = sum(score_by_month[person].values())
    score_count = len(score_by_month[person])
    avg_score = round(score_total / score_count)

    #add score to dictionary 
    individual_average[person] = avg_score

print(f"Dictionary of the average credit score of each individual: \n{individual_average}")


                  
```

```python
def get_int(min, max):
    try:
        num = input(f"Please Input an integer between {min} and {max}: ")
        if "." in num:
            raise Exception("float, not int")
        num = int(num)
        
    except:
        print("that was not an integer value")
        num = 0
    else:
        try:
            if num <= min or num >= max:
                raise Exception("Number out of Range")
                #are we supposed to divide by zero to get an error? why?
        except:
            print("Please input an integer in the correct range")
        else: 
            return num
```
